## making requests
exercise on API's

```python
import json, requests

url = "https://api.disneyapi.dev/character"
response = requests.get(url)
# reponse = requests.Request("GET", url) --> old way of doing requests

text = response.text
data = json.loads(text)

# check which char has the most movies
counter = 0
for char in data["data"]:
    if len(char["films"]) > counter:
        counter = len(char["films"])
        s = char["name"]

print(f"The following character has the most movies {s} with {counter} movies!")


#find characters with at least 1 movie and 1 videogame

print()

L = []
for char in data["data"]:
    if char["videoGames"] != [] and char["films"] != []:
        L.append(char["name"])

print("These are all the characters that are both in videogames as well as in films: ")
for item in L:
    print(item)
```

## WEBSCRAPING

```python
import requests, re

#I want to look for elephants!

url = "https://en.wikipedia.org/wiki/Elephant"

headers = {
    'User-Agent': 'Any name'
}
response = requests.get(url, headers=headers)

if response.status_code == 200:
    html = response.text
    regex = "Elephant|elephant"
    matches = re.findall(regex, html)
    print(len(matches))
else:
    print(response.text)
```

```python
#WORKING WITH BEAUTIFULSOUP
import requests 
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Elephant"
headers = {'User-Agent': ''}
response = requests.get(url, headers=headers)
html = response.text
soup = BeautifulSoup(html)
#print(soup) --> in case you want to see content of a page!
#print(soup.prettify)

print(soup.title)
print(soup.title.get_text())
print(soup.get_text()[0:500].split())
print(soup.get_text().count("Elephant"))
print(soup.get_text().count("elephant"))

h3_tags = soup.find("h3") 
print(h3_tags)
h3_tags = soup.find_all("h3")
print(h3_tags)

for element in h3_tags:
    print(element.get_text())
```

```python
#getting all language from elephantis

import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Elephant"
headers = {'User-Agent': ''}
response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text)
li_tags = soup.find_all("li",{'class':"interlanguage-link"})

for tag in li_tags:
    T = tag.a["title"].split()
    print(f"{T[0]} is elephant in {T[2]}")
```

```python
#find some images and regex
import requests
from bs4 import BeautifulSoup
import re

url = "https://en.wikipedia.org/wiki/Elephant"
headers = {'User-Agent': ''}
response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text)
img_tags = soup.find_all("img") #all img tags
src = [img["src"] for img in img_tags]

#below is same as the oneliner above here!
#src = []
#for img in img_tags:
#    src.append(img["src"])
print(src)
result = re.findall(r'(\w+[.](jpg|png|gif))', str(src))
count = 0

for source in result:
    if "wikimedia" in source:
        print(source)
        filename = "Elephants/" + str(count) + ".jpg"
        r = requests.get("https:" + source[0])
        with open(filename, 'wb') as file:
            file.write(r.content)
        count += 1

```
