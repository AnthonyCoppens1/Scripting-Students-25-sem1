# MULTITHREADING
python code tends to always run single threaded, meaning 1 process at a time to be executed

```python
import time
def sleep():
    time.sleep(5)
    print("WAKE THE FUCK UP")

def calc():
    for i in range(0, 10):
        print(i * 5)


#single threaded:
print("START")
sleep()
calc()
print("STOP")
```

```python
#MULTITHREADED
import time
def sleep():
    time.sleep(5)
    print("WAKE THE FUCK UP")

def calc():
    for i in range(0, 10):
        print(i * 5)

thread_sleep = threading.Thread(target=sleep)
thread_calc = threading.Thread(target=calc)

print("START")
thread_sleep.start()
thread_calc.start()
print("STOP")
```

```python
import threading, time
#threading or multithreading
def function1():
    for i in range(1,11):
        print(i, end=" ")
        time.sleep(0.5)

def function2():
    for c in "ABCDEFGHIJ":
        print(c, end=" ")
        time.sleep(0.51)

thread_function1 = threading.Thread(target=function1)
thread_function2 = threading.Thread(target=function2)

print("START")
thread_function1.start()
thread_function2.start()
thread_function1.join()
thread_function2.join()
print("STOP")
```

```python
#working with date-formats
import datetime as dt
print(dt.datetime.now())

date = dt.datetime(2000, 5, 10, 0, 0, 0)
print(date)
date = dt.datetime.fromtimestamp(time.time())

print(date.year, date.month, date.day, date.hour, date.minute, date.second)
print(dt.datetime.fromtimestamp(1763105327))
print(dt.datetime.fromtimestamp(time.time()))
print(dt.datetime.fromtimestamp(round(time.time(), 0)))

#going from a datetime into a string --> strftime()
print(date.strftime("%d %B %y"))
print(date.strftime("day %j of %Y"))

date += dt.timedelta(days=1000)
print(date.strftime("%d %B %y"))
```
