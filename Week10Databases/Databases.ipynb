```python
import mysql.connector

dbconnection = mysql.connector.connect(
    host = "localhost",
    #port = "3307" --> FOR THE PEOPLE THAT CHANGED PORT
    user = "root",
    password = "", #IF YOU CHANGED PORT? THERE IS NO PW LINKED TO THAT NEW PORT!
    database = "movies_db"
)

cursor = dbconnection.cursor()
query = "INSERT INTO movies(title, synopsis) VALUES(%s, %s);"

data = [("Oppenheimer", "BIG BOMBA"), ("Scarface", "Gangster"), ("The dictator", "Alladeen trying to push a point bomb"),
        ("The imitation game", "Decrypts stuff"), ("Jurrasic Park", "Random dino's going rampage"), ("The mask", "Looney Loki")]

#insert all of the movies into the database
cursor = dbconnection.cursor()
for movie in data:
    cursor.execute(query, movie)

#also possible: cursor.executemany(query, data)
dbconnection.commit()

cursor.close()
dbconnection.close()
```

## READING FROM DATABASES --> select

```python
import mysql.connector

dbconnection = mysql.connector.connect(
    host = "localhost",
    user = "root",
    password = "",
    #port = "3307" --> FOR THE PEOPLE THAT CHANGED PORT
    database = "movies_db"
)

cursor = dbconnection.cursor()

query = "SELECT  * FROM actors;"
cursor.execute(query)

data = cursor.fetchall()
print(data)

for actor in data:
    print(actor)

for id, actor in data:
    print(actor)

cursor.close()
dbconnection.close()
```
